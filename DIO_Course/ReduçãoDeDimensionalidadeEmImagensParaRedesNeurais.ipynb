```python
#OBS:  O codigo foi feito para rodar no colab, se for rodar local tem que trocar o
# cv2_imshow()
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

image_path = "/content/drive/MyDrive/Competicoes/Dio/space.jpg"

print("Imagem original")
image = cv2.imread(image_path)
cv2_imshow(image)

print("Imagem em Cinza")
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)

print("Imagem Binaria")
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
thresholded_image = np.where(image < 127, 1, 0)
cv2_imshow(thresholded_image*255)
```
